# Explanation about the features

1. **symbolling** - corresponds to the insurance risk level of a car. Cars are initially assigned a risk factor symbol associated with their price. Then, if an automobile is riskier, this symbol is adjusted by moving it up the scale. A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe. <br>

2. **normalized losses** - is the relative average loss payment per insured vehicle year. This value is normalized for all autos within a particular size classification (two-door, small, station wagons, sports/specialty, etc…), and represents the average loss per car per year. The values range from 65 to 256.<br>

3. **make** - is the brand of the vehicle (for example: Toyota, Mazda, etc).<br>

4. **fuel type** - is the type of fuel that each vehicle consumes. In our dataset two types are present: (gas/gasoline and diesel). Having in mind that the data about cars in 1985 these two were commonly available everywhere back then.<br>

5. **aspiration** - is the type of aspiration that the engine can be. A naturally-aspirated engine is one that is not fitted with a turbocharger or supercharger. In other words, air goes into the engine normally, whereas turbocharged and supercharged engines force the air in under pressure to boost power. In our dataset we have the two types: std (standard or naturally-aspirated) and turbo (turbocharched engine).<br>

6. **num-of-doors** - this feature corresponds to the number of accesses the passenger has to the vehicle. In our dataset two types can be noted: two and four. Four door cars can be sometimes mentioned as 5-door cars because of the trunk in the back but our dataset has the 4-door definition.<br>

7. **body-style** - is the style of the vehicle. Five types can be observed in this dataset: hatchback, sedan, wagon, convertible and hardtop. *Convertible* has a retractable or removeable roof. A convertible allows an open-air driving experience, with the ability to provide a roof when required. Most convertible roofs are either a folding textile soft-top or a retractable metal roof. *Sedan* or fixed-roof is car in a three-box design with separate compartments for engine, passenger, and cargo. Sedans have 4 doors. A sedan is called a "berlina" in Spanish and Italian, or a "berline" in French. A *hatchback* is a car body configuration with a rear door that swings upward to provide access to a cargo area. A station wagon, also called an estate (UK) or simply *wagon* (US), is body-style type of a sedan/saloon with its roof extended rearward over a shared passenger/cargo volume with access at the back via a third or fifth door, instead of a trunk lid. *Hardtop*  is any car with a rigid roof.<br>

8. **drive-wheels** -  there are four different types of drivetrain: all-wheel-drive (AWD), front wheel drive (FWD), rear wheel drive (RWD), and 4WD (4 wheel drive). *FWD* means that the power from the engine is delivered to the front wheels of your vehicle. With FWD, the front wheels are pulling the car and The rear wheels don’t receive any power on their own. *RWD* means that the power from the engine is delivered to the rear wheels and the rear wheels push the car forward. The front wheels do not receive any power and are free to manoeuvre the vehicle. *4WD* means the power from the engine is delivered to all 4 wheels all of the time 4X4 is engaged and has an option to operate in a RWD format to conserve fuel.

9. **engine-location** -  As the name says, it tells us where is the engine located: back or front.

10. **wheel-base** - is the horizontal distance between the centers of the front and rear wheels (in inches).

11. **length** - is the length of the vehicle. Again, in this dataset the measures are in inches. For example 168.8 inches equals 4.29 meters.

12. **width** - is the width of the vehicle. Also, this attribute's values are in inches. 

13. **height** - is the measure how tall is the vehicle. Also, this attribute's values are in inches.

14. **curb-weight** - is the weight of the vehicle without any passenger or items in it except for the standard equipment that comes with it. This attribute's values are measured in lbs (pound).

15. **engine-type** - states how the engine is assembled or designed in terms of operations of valves and cylinders. In this dataset we have seven engine types: *dohc* (Dual OverHead Cam), *dohcv* (Dual OverHead Cam and Valve), *l* (L engine), *ohc* (OverHead Cam), *ohcf* (OverHead Cam and Valve F engine), *ohcv* (OverHead Cam and Valve) and *rotor* (Rotary engine).

16. **num-of-cylinders** - the number of cylinders of a car is related to how the engine works, where generally more cylinders mean more potential power. In this dataset we have cylinders in the range from 2 to 6, 8 and 12.

17. **engine-size** - engine size is the volume of fuel and air that can be pushed through a car's cylinders and is measured in cubic centimetres (cc). In this dataset it is measured in cubic inches (ci).

18. **fuel-system** - the fuel system is made up of the fuel tank, pump, filter, and injectors or carburetor, and is responsible for delivering fuel to the engine as needed. In the dataset, 8 types are present: *mpfi (Multi Point Fuel Injection)*, *2bbl (2 barrel carb)*, *mfi Multi-Port Fuel Injection, 1bbl (1 barrel carb), spfi (Spark Plug Fuel Injector), 4bbl (4 barrel carb), idi(Indirect fuel injection), spdi (spatial plasma discharge ignition).*

19. **bore** - in a engine, the bore (or cylinder bore) is the inside diameter of each cylinder. Also, this attribute is measured in inches.

20. **stroke** - is the stroke length how far the piston travels in the cylinder. This attribuite is measured in inches.

21. **compression-ratio** - in an internal-combustion engine, is degree to which the fuel mixture is compressed before ignition. Compression ratios usually range from 8:1 to 10:1. A higher compression ratio, say from 12:1 to 14:1 means higher combustion efficiency.

22. **horsepower** - (hp) is the metric used to indicate the power produced by a car's engine - the higher the number, the more power is sent to the wheels and, in theory, the faster it will go. 

23. **peak-rpm** - peak torque is the maximum torque that a machine or motor can produce to achieve a given number of revolutions per minute, or rpm.

23. **city-mpg** - or miles per gallon, is the distance measured in miles, that a car can travel per gallon of fuel in city conditions, with stopping and starting at lower speeds. Mpg is also the primary measurement of a car's fuel efficiency: The higher a car's mpg, the more fuel efficient it is.

24. **highway-mpg** - is the distance measured in miles, that a car can travel per gallon of fuel  while driving on an open stretch of road without stopping or starting, typically at a higher speed.

25. **price** - is the price of the car in EUR.

### 1.2. Handling missing values

One way to deal with missing values is to fill them with the **mean or the median**. At this point, we don't know whether there are outliers in the dataset. What we know is that **the mean is very sensitive to outliers and the median isn't affected by extreme values.**

We will compare the mean and the median of the numerical columns with missing values. <br>
On the first hand, if the mean and the median are **fairly close**, we know that the data has a **symmetrical distribution** (the data is fairly balanced, or symmetric, on each side).

On the other hand, if they are **significantly different**, that tells us that the data is probably **skewed** in one direction or another and could affect our decisions. In a data distribution with **extreme outliers**, the distribution is skewed in the direction of the outliers.

If the **mean is much smaller than the median**, the data is generally **skewed left**. If the **mean is much larger than the median**, the data is generally **skewed right**.

```python
attributes = ['bore', 'stroke', 'horsepower', 'peak-rpm', 'normalized-losses']

for a in attributes:
    df[a] = df[a].astype(float) #needed because of the missing values
    print(f'The mean of the attribute {a} is {df[a].mean()}.')
    print(f'The median of the attribute {a} is {df[a].median()}.')
    print('_____________________________________________________________\n')
```

From the function above, we can see that the mean and the median of the attributes **bore, stroke** and **normalized-losses** seem to be fairly close, which means that deciding whether to replace the NaNs with the mean or with the median won't make a big difference. <br>For these attributes, we will impute the missing values with the mean of each attribute.

The mean of the attribute **horsepower** is significantly larger than the median, which means that the data is right-skewed and there might be outliers. <br>
In contrast to this, the mean of the attribute **peak-rpm** is smaller than the median, which means that the data is left-skewed and might contain outliers. <br>We will impute the missing values for these attributes with their median value, since the median is not affected by outliers.

```python
df['bore'].fillna((df['bore'].mean()), inplace=True)
df['stroke'].fillna((df['stroke'].mean()), inplace=True)

#Visualizing the distribution of the attribute normalized-losses before and after imputing the missing values
sns.kdeplot(df['normalized-losses'])
plt.title('Before replacing nan values with the average of the column')
plt.show()

df['normalized-losses'].fillna((df['normalized-losses'].mean()), inplace=True)

sns.kdeplot(df['normalized-losses'])
plt.title('After replacing nan values with the average of the column')
plt.show()
```

#### 2.2.5. Handling the compression-ratio outliers

The number of compression-ratio outliers is relatively high (20). Therefore, we did a small research and found out that diesel engines use higher compression ratios than petrol engines, because the lack of a spark plug means that the compression ratio must increase the temperature of the air in the cylinder sufficiently to ignite the diesel using compression ignition. The function **calc_vif** calculates the VIF value for each attribute and creates a dataframe where one column is the attribute itself and the other is its VIF value.

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(dataset):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["features"] = dataset.columns
    vif["VIF"] = [variance_inflation_factor(dataset.values, i) for i in range(dataset.shape[1])]

    return(vif)
```

The function **check_vif_corr** calls the previously defined function calc vif. 

```python
def check_vif_corr(df_to_drop_from, attribute_to_drop):
    features_dropped = df_to_drop_from.drop([attribute_to_drop], axis=1) #dropping the specified attribute from the specified dataset
    datasets.append(features_dropped) #appending the new dataset to the list of datasets
    features = features_dropped.select_dtypes(include = ["float64","int64"]).iloc[:,:-1] #selecting all of the numerical variables except for the dependent one
    print(calc_vif(features)) #calling the calc_vif function with the numerical attributes
```

### 3.2. RFE Feature Selection

Recursive feature elimination (RFE) is a feature selection method that fits a model and removes the weakest feature (or features) until the specified number of features is reached. It does not consider multicollinearity and significance levels of the attributes.

First, we split the dataset. Then, we scale the data with MinMaxScaler().

```python
def rfe_feature_selection(num_features):
    model = ExtraTreesRegressor()
    rfe = RFE(model, num_features)
    fit = rfe.fit(X_train, Y_train)
    pred = fit.predict(X_test)

    print("Number of Features: ", num_features)
    print("Selected Features: ", fit.support_)
    print("Feature Ranking: ", fit.ranking_) 
    print("====================================================================================================")
    print("MAE:",mean_absolute_error(Y_test,pred))
    print("MSE:",mean_squared_error(Y_test,pred))
    print("R^2:  {:.5f}".format(fit.score(X_test, Y_test)))
    print("____________________________________________________________________________________________________\n\n")
```

### 3.4. Regularization

Trying to penalize the loss function with different regression algorithms such as Ridge Regression, Lasso Regression, Elastic Net (combining the L1 and L2 penalty functions) and Bayesian Ridge Regression. The model variance and the obtained MSE, RMSE, MAE and RMAE are compared in the table below.

The BayesianRidge results with the best MSE, RMSE, MAE and R2. Lasso and ElasticNet give identical results.

```python
algorithms = []
algorithms.append(Ridge(alpha=1))
algorithms.append(Lasso(alpha=0.1))
algorithms.append(ElasticNet(alpha=0.1, l1_ratio = 0.4))
algorithms.append(BayesianRidge())

data = []
for algo in algorithms:
    algo.fit(X_train, Y_train)
    y_pred = algo.predict(X_test)
    sigma = np.sum((Y_test - algo.predict(X_test))**2) / (X_train.shape[0]-2)
    data.append(({
        'MSE': mean_squared_error(Y_test, y_pred),
        'RMSE': np.sqrt(mean_squared_error(Y_test, y_pred)),
        'MAE': mean_absolute_error(Y_test, y_pred),
        'R2': r2_score(Y_test, y_pred),
        'model variance': sigma
    }))
    
    
results = pd.DataFrame(data=data, columns=['MSE', 'RMSE', 'MAE', 'R2', 'model variance'],
                       index=['Ridge', 'Lasso', 'ElasticNet', 'BayesianRidge'])

results
```

After preforming a box-cox transformation on the dependent variable, we notice that the area in which we can expect the results to be is much smaller than before. This means that our model's predictive power is high and the prediction error is low.

The lasso model gives the best results for alpha = ~ 10^(-3.5).

```python
numClusters = 7
SSE = []
silhouette_coefficients = []

for k in range(1, numClusters):
    k_means = KMeans(n_clusters = k)
    k_means.fit(df_robust)
    SSE.append(k_means.inertia_)
    if k<=1:
        continue;
    score = silhouette_score(df_robust, k_means.labels_)
    silhouette_coefficients.append(score)
    
plt.plot(range(1, numClusters), SSE);
plt.xlabel('Number Of Clusters');
plt.ylabel('SSE');
```

```python
kl = KneeLocator(range(1, numClusters), SSE, curve="convex", direction="decreasing")
kl.elbow
```

The Silhouette Visualizer displays the silhouette coefficient for each sample on a per-cluster basis, visually evaluating the density and separation between clusters. The score is calculated by averaging the silhouette coefficient for each sample, computed as the difference between the average intra-cluster distance and the mean nearest-cluster distance for each sample, normalized by the maximum value. This produces a score between -1 and +1, where scores near +1 indicate high separation and scores near -1 indicate that the samples may have been assigned to the wrong cluster.

In SilhouetteVisualizer plots, clusters with higher scores have wider silhouettes, but clusters that are less cohesive will fall short of the average score across all clusters, which is plotted as a vertical dotted red line.
